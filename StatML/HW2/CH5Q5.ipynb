5c (combined 5b into 1 cell to repeat 3 times)

```python
def calculate_error():
  # bi
  # split the dataset into training and validation sets, here
  train = Default.sample(frac=0.5).sort_index()
  val = Default.drop(train.index)

  # ground truths
  ytrain = train['default']
  ytrain = ytrain.map(lambda x: 0 if x == 'No' else 1)
  yval = val['default']
  yval = yval.map(lambda x: 0 if x == 'No' else 1)

  # x data
  xtrain = train[['balance', 'income']]
  xval = val[['balance', 'income']]

  # bii
  glm = sm.GLM(ytrain, xtrain, family=sm.families.Binomial())
  results = glm.fit()

  # biii
  preds = results.predict(xval)
  preds = preds.map(lambda x: 0 if x < 0.5 else 1)

  # biv
  error = np.mean(preds != yval)
  return error
```

```python
[calculate_error() for _ in range(3)]
```

There seems to be minimal difference in the error rates when using different splits.

```python
# reuse calculate_error() defined in 5c; just add dummy variable student in x
def calculate_error_withdummy():
  # for student and default, change yes/no to 1/0
  data = Default.copy() # prevent data corruption
  data['student'] = data['student'].map(lambda x: 0 if x == 'No' else 1)
  data['default'] = data['default'].map(lambda x: 0 if x == 'No' else 1)

  # bi
  # split the dataset into training and validation sets, here
  train = data.sample(frac=0.5).sort_index()
  val = data.drop(train.index)

  # ground truths
  ytrain = train['default']
  yval = val['default']

  # x data
  xtrain = train[['balance', 'income', 'student']]
  xval = val[['balance', 'income', 'student']]

  # bii
  glm = sm.GLM(ytrain, xtrain, family=sm.families.Binomial())
  results = glm.fit()

  # biii
  preds = results.predict(xval)
  preds = preds.map(lambda x: 0 if x < 0.5 else 1)

  # biv
  error = np.mean(preds != yval)
  return error
```

```python
[calculate_error_withdummy() for _ in range(3)]
```

Adding a dummy variable does not really affect the test error rate.
